```python
import pandas as pd
import os.path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(os.path.join(os.path.expanduser('~'), 'Downloads/FMT_data.xlsx'), index_col=0, header=[0, 1])
```

```python
df.head()
```

```python
df.columns = df.columns.rename("DAYS", level=0)
df.columns = df.columns.rename("PATIENT", level=1)
```

```python
df.head()
```

```python
df.columns[0]
```

```python
df.columns[1]
```

```python
df2 = df.T
```

```python
df2.head()
```

```python
df2['Mean'] = df2.iloc[:, 0:].mean(axis=1)
```

```python
df2.head()
```

```python
df2 = df2.reset_index()
```

```python
df2.head()
```

```python
colors = ['#78C850', '#F08030', '#6890F0', 
          '#A8B820', '#A8A878', '#A040A0', 
          '#F8D030', '#E0C068', '#EE99AC', 
          '#C03028', '#F85888', '#B8A038', 
          '#705898']
```

```python
plt.figure(figsize=(4, 8))
sns.set(style="ticks", color_codes=True)
ax = sns.swarmplot(data=df2, x='DAYS', y='Mean', hue='PATIENT',
                   size=12, palette=colors)

sns.boxplot(data=df2, x="DAYS", y="Mean",
                 showcaps=False,boxprops={'facecolor':'None'},
                 ax=ax)
plt.legend(bbox_to_anchor=(1, 1), loc=2);

plt.show()
```

```python
df2.iloc[:,:-1].head()
```

```python
df3 = df2.iloc[:,1:-1].reset_index()
```

```python
df3.head()
```

```python
df3 = df3.set_index('PATIENT')
```

```python
df3.drop(['index', 'UNMAPPED', 'UNGROUPED'], axis=1, inplace=True)
```

```python
df3.head()
```

```python
df3 = df3.set_index('PATIENT')
```

```python
df3.head()
```

```python
g = sns.clustermap(df3, cmap="Blues", method="average", metric="correlation",
               col_cluster=True, row_cluster=True, z_score=0, figsize=(12, 18));
```

```python
g.savefig("heatmap_corr_zscore.png")
```
